```python
import matplotlib.pyplot as plt
%matplotlib inline

x = [i for i in range(101)]
y1 = [i ** 2 for i in x]
plt.plot(x, y1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3,50)
y1 = 2 * x + 1
y2 = x ** 2

plt.figure()
plt.plot(x, y1)

plt.figure(num = 5,figsize = (8, 5))
plt.plot(x, y2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X) # for color value
plt.scatter(X, Y, s = 75, c = T, alpha = 0.5)

plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

for x,y in zip(X, Y1):
    
    #ha: horizontal alignment
    plt.text(x + 0.2, y + 0.01, '%.2f' % y, ha = 'center', va = 'bottom')

plt.bar(X, Y1)
plt.show()
```
